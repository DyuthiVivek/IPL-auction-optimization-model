```python
import pandas as pd
```

```python
data = pd.read_csv("final.csv")
```

```python
data.head(10)
```

```python
overseas_indices = data[data['Player Origin'] == 'Overseas']['Unnamed: 0']
```

```python
player_variables = ["x" + str(i) for i in range(len(data))]
    
add_variables = "\n".join([f'{var} = model.addVar("{var}", vtype="BINARY")' for var in player_variables])
print(add_variables)
```

```python
print("model.addCons(" + " + ".join(player_variables[:110]) + " == v1)")
print("model.addCons(" + " + ".join(player_variables[110:220]) + " == v2)")
print("model.addCons(" + " + ".join(player_variables[220:]) + " == v3)")
```

```python
print("model.addCons(" + " + ".join(["x" + str(ind) for ind in overseas_indices]) + " <= 4)")
```

```python
performance = data['average_bowling_performance_index'] + data['average_batting_performance_index']
perf_variables = list(zip(performance, player_variables))

print("model.addCons(" + " + ".join([f'{p} * {var}'for p, var in perf_variables[:110]]) + " == v4)")
print("model.addCons(" + " + ".join([f'{p} * {var}'for p, var in perf_variables[110:220]]) + " == v5)")
print("model.addCons(" + " + ".join([f'{p} * {var}'for p, var in perf_variables[220:]]) + " == v6)")
```

```python
bowling_scores = list(zip(data['average_bowling_performance_index'], player_variables))

print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in bowling_scores[:110]), "== b1)")
print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in bowling_scores[110:220]), "== b2)")
print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in bowling_scores[220:]), "== b3)")
```

```python
batting_scores = list(zip(data['average_batting_performance_index'], player_variables))

print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in batting_scores[:110]), "== b4)")
print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in batting_scores[110:220]), "== b5)")
print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in batting_scores[220:]), "== b6)")
```

```python
cost = list(zip(data['Amount'], player_variables))

print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in cost[:110]), "== u1)")
print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in cost[110:220]), "== u2)")
print("model.addCons(", " + ".join(f'{score} * {var}' for score, var in cost[220:]), "== u3)")
```

```python
performance = data['average_bowling_performance_index'] + data['average_batting_performance_index']
perf_variables = list(zip(performance, player_variables))

print("model.addCons(" + " + ".join([f'{var} * ({p} - pmean) ** 2' for p, var in perf_variables[:110]]) + " == p1)")
print("model.addCons(" + " + ".join([f'{var} * ({p} - pmean) ** 2' for p, var in perf_variables[110:220]]) + " == p2)")
print("model.addCons(" + " + ".join([f'{var} * ({p} - pmean) ** 2' for p, var in perf_variables[220:]]) + " == p3)")
```
